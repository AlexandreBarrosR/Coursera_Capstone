```python
response = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M");
```

#### Using BeautifulSoup to scrapping the html returned, retrieve the table and transform into a dataframe

```python
soup = BeautifulSoup(response.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))[0]
df
```

#### Removing borough "Not assigned" cells

```python
df = df[df["Borough"] != "Not assigned"].reset_index(drop=True)
df
```

#### Getting Geospatial data from csv

```python
df_geolocations = pd.read_csv('http://cocl.us/Geospatial_data')
df_geolocations.head()
```

```python
latSeries = []
longSeries = []

for pcode in df["Postal code"]:
    geolocation = df_geolocations[df_geolocations["Postal Code"] == pcode]
    latSeries.append("{0:.6f}".format(float(geolocation["Latitude"])))
    longSeries.append("{0:.6f}".format(float(geolocation["Longitude"])))
    
df["Latitude"] = latSeries
df["Longitude"] = longSeries
df
```

```python
df_to_explore = df
df_to_explore.head()
```

#### Using Foursquare to explore Borough York

```python
VERSION = '20180605'
LIMIT = 100
radius = 500

venues_list=[]

for name, lat, lng in zip(df_to_explore["Neighborhood"], df_to_explore["Latitude"], df_to_explore["Longitude"]):
    print(name)
    # create the API request URL
    url = 'https://api.foursquare.com/v2/venues/explore?&oauth_token=QCYA5YYIDY2KAMAOWD5AJM20U4QGMVGXYEEQ1OGYFYU2QYAx&v={}&ll={},{}&radius={}&limit={}'.format(
        VERSION, 
        lat, 
        lng, 
        radius, 
        LIMIT)

    # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
    # return only relevant information for each nearby venue
    venues_list.append([(
        name, 
        lat, 
        lng, 
        v['venue']['name'], 
        v['venue']['location']['lat'], 
        v['venue']['location']['lng'],  
        v['venue']['categories'][0]['name']) for v in results])

nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
nearby_venues.columns = ['Neighborhood', 
              'Neighborhood Latitude', 
              'Neighborhood Longitude', 
              'Venue', 
              'Venue Latitude', 
              'Venue Longitude', 
              'Venue Category']

nearby_venues
```

#### Check how many distincts Venue categories

```python
print('There are {} uniques categories.'.format(len(nearby_venues['Venue Category'].unique())))
```

```python
nearby_venues_onehot = pd.get_dummies(nearby_venues[['Venue Category']], prefix="", prefix_sep="")
nearby_venues_onehot
```

```python
nearby_venues_onehot['Neighborhood'] = nearby_venues['Neighborhood']
fixed_columns = [nearby_venues_onehot.columns[-1]] + list(nearby_venues_onehot.columns[:-1])
nearby_venues_onehot = nearby_venues_onehot[fixed_columns]

nearby_venues_onehot.head(15)
```

#### Getting the Venue Categorie frequency for each Neighborhood

```python
nearby_venues_grouped = nearby_venues_onehot.groupby('Neighborhood').mean().reset_index()
nearby_venues_grouped.head()
```

#### Getting top10 venues for each neighborhood

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    columns.append('#{} Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = nearby_venues_grouped['Neighborhood']

for ind in np.arange(nearby_venues_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(nearby_venues_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

#### Clustering neighborhood

```python
kclusters = 5

nearby_venues_grouped_clustering = nearby_venues_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(nearby_venues_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

nearby_venues_merged = nearby_venues

nearby_venues_merged = nearby_venues_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

nearby_venues_merged.head() # check the last columns!
```

#### Getting Toronto coordinates to positioning the map

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of {} are {}, {}.'.format(address, latitude, longitude))
```

#### Creating map to visualize clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(nearby_venues_merged['Neighborhood Latitude'], nearby_venues_merged['Neighborhood Longitude'], nearby_venues_merged['Neighborhood'], nearby_venues_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [float(lat), float(lon)],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
